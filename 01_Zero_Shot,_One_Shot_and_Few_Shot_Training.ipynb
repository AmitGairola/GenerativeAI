```python
from datasets import load_dataset
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer, GenerationConfig, TrainingArguments, Trainer
import torch
import time
import pandas as pd
import numpy as np
```

```python
dataset_name = 'knkarthick/dialogsum'

ds = load_dataset(dataset_name)
```

```python
sample_indices = [10,20,30]

dash_line = '-'.join('' for x in range(100))


for i, index in enumerate(sample_indices):
  print(dash_line)
  print('Example ',i+1)
  print(dash_line)
  print('INPUT DIALOGUE: ')
  print(ds['test'][index]['dialogue'])
  print(dash_line)
  print('BASELINE HUMAN SUMMARY: ')
  print(ds['test'][index]['summary'])
  print(dash_line)
  print()
```

#Zero Shot Inference without Prompt Engineering

```python
model_name = 'google/flan-t5-base'

model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
```

```python
def print_summary(i,index, dialog, summary, output=""):
  print(dash_line)
  print("Example ",i+1)
  print(dash_line)
  print("INPUT DIALOGUE: ")
  print(ds['test'][index]['dialogue'])
  print(dash_line)
  print("BASELINE HUMAN SUMMARY :")
  print(ds['test'][index]['summary'])
  print(dash_line)

  if output !="":
    print("MODEL GENERATED SUMMARY :")
    print(output)
    print(dash_line)
  print()
```

```python
for i, index in enumerate(sample_indices):
  dialog = ds['test'][index]['dialogue']
  summary = ds['test'][index]['summary']

  inputs = tokenizer(dialog, return_tensors='pt') # Check for why return tensors = pt is needed
  predict = model.generate(inputs['input_ids'], max_new_tokens = 50)[0]
  output = tokenizer.decode(predict, skip_special_tokens=True)

  print_summary(i, index, dialog, summary, output)
```

# Prompt Engineering

```python
for i, index in enumerate(sample_indices):
  dialog = ds['test'][index]['dialogue']
  summary = ds['test'][index]['summary']

  prompt = f""" Summarize the following conversation. {dialog} Summary: """

  inputs = tokenizer(prompt, return_tensors='pt')
  predict = model.generate(inputs['input_ids'], max_new_tokens=100)[0]
  outputs = tokenizer.decode(predict, skip_special_tokens=True)

  print_summary(i,index, dialog, summary, outputs)
```

# **Using Prompt template from FLAN-T5 and Zero Shot Inference**

```python
for i, index in enumerate(sample_indices):
  dialog = ds['test'][index]['dialogue']
  summary = ds['test'][index]['summary']

  prompt = f""" Dialogue: {dialog}

  What was going on? """

  inputs = tokenizer(prompt, return_tensors='pt')
  predict = model.generate(inputs['input_ids'], max_new_tokens=100)[0]
  outputs = tokenizer.decode(predict, skip_special_tokens=True)

  print_summary(i, index, dialog, summary, outputs)
```

```python
example_index = 40

prompt_in = f""" Dialogue
{ds['test'][example_index]['dialogue']}

What was going on?
{ds['test'][example_index]['summary']}
"""

for i, index in enumerate(sample_indices):
  dialog = ds['test'][index]['dialogue']
  summary = ds['test'][index]['summary']

  prompt = prompt_in + f""" Dialogue {dialog}
  What was going on"
  """

  inputs = tokenizer(prompt, return_tensors='pt')
  predict = model.generate(inputs['input_ids'], max_new_tokens=100)[0]
  outputs = tokenizer.decode(predict, skip_special_tokens=True)

  print_summary(i,index, dialog, summary, outputs)
```

# ***Few Shot Inference***

```python
few_shot_examples = [50,600]

for a, index in enumerate(few_shot_examples):
  d = ds['test'][index]['dialogue']
  s = ds['test'][index]['summary']
  prompt_in += f""" Dialogue :
  {d} \n
  What was going on?\n
  {s}\n
  """

for i, index in enumerate(sample_indices):
  dialog = ds['test'][index]['dialogue']
  summary = ds['test'][index]['summary']

  prompt = prompt_in +"\n" + f""" Dialogue :
  {dialog} \n
  What was going on? \n
  """

  inputs = tokenizer(prompt, return_tensors = 'pt')
  predict = model.generate(inputs['input_ids'], max_new_tokens=100)[0]
  outputs = tokenizer.decode(predict, skip_special_tokens=True)

  print_summary(i, index, dialog, summary, outputs)
```
